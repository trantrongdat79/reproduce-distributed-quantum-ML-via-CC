```python
import numpy as np
import time
import pickle
import matplotlib.pyplot as plt
def make_array(lst):
    result=[]
    for item in lst:
        result.append(np.array(item))
    return(np.array(result))
```

```python
CC=[]
non=[]
QC=[]
NC=[]

for i in range(1,11):
    with open('CCDQML_depth9_Dataset3_'+str(i), 'rb') as f:
        CC_tmp=pickle.load(f) 
        CC.append(CC_tmp)

for i in range(1,11):
    with open('nonDQML_depth9_Dataset3_'+str(i), 'rb') as f:
        non_tmp=pickle.load(f) 
        non.append(non_tmp)

for i in range(1,11):
    with open('QCDQML_depth9_Dataset3_'+str(i), 'rb') as f:
        QC.tmp=pickle.load(f) 
        QC.append(QC.tmp)

for i in range(1,11):
    with open('NCDQML_depth9_Dataset3_'+str(i), 'rb') as f:
        NC_tmp=pickle.load(f) 
        NC.append(NC_tmp)

CC=np.array(CC)
non=np.array(non)
QC=np.array(QC)
NC=np.array(NC)
Accuracy_CC=CC[:,1]
Accuracy_non=non[:,1]
Accuracy_QC=QC[:,1]
Accuracy_NC=NC[:,1]
Accuracy_CC=make_array(Accuracy_CC)
Accuracy_non=make_array(Accuracy_non)
Accuracy_QC=make_array(Accuracy_QC)
Accuracy_NC=make_array(Accuracy_NC)
meanA_CC=np.mean(Accuracy_CC,axis=0)[:500]
meanA_non=np.mean(Accuracy_non,axis=0)[:500]
meanA_QC=np.mean(Accuracy_QC,axis=0)[:500]
meanA_NC=np.mean(Accuracy_NC,axis=0)[:500]
stdA_CC=np.std(Accuracy_CC,axis=0)[:500]
stdA_non=np.std(Accuracy_non,axis=0)[:500]
stdA_QC=np.std(Accuracy_QC,axis=0)[:500]
stdA_NC=np.std(Accuracy_NC,axis=0)[:500]
plt.fill_between(range(1,501),meanA_non-stdA_non,meanA_non+stdA_non,alpha=0.2,color='g')
plt.plot(range(1,501),meanA_non,label=f'nonDQML',color='g')
plt.fill_between(range(1,501),meanA_NC-stdA_NC,meanA_NC+stdA_NC,alpha=0.2,color='darkorange')
plt.plot(range(1,501),meanA_NC,label=f'NCDQML',color='darkorange')
plt.fill_between(range(1,501),meanA_CC-stdA_CC,meanA_CC+stdA_CC,alpha=0.2,color='r')
plt.plot(range(1,501),meanA_CC,label=f'CCDQML',color='r')
plt.fill_between(range(1,501),meanA_QC-stdA_QC,meanA_QC+stdA_QC,alpha=0.2,color='b')
plt.plot(range(1,501),meanA_QC,label=f'QCDQML',color='b')

meanA_CC=np.mean(Accuracy_CC,axis=0)
meanA_non=np.mean(Accuracy_non,axis=0)
meanA_QC=np.mean(Accuracy_QC,axis=0)
meanA_NC=np.mean(Accuracy_NC,axis=0)
stdA_CC=np.std(Accuracy_CC,axis=0)
stdA_non=np.std(Accuracy_non,axis=0)
stdA_QC=np.std(Accuracy_QC,axis=0)
stdA_NC=np.std(Accuracy_NC,axis=0)
plt.legend(loc=4)
plt.savefig('IterAcc_dp9.pdf')
plt.show()

A_non_9=meanA_non[999]
s_non_9=stdA_non[999]
A_CC_9=meanA_CC[999]
s_CC_9=stdA_CC[999]
A_NC_9=meanA_NC[999]
s_NC_9=stdA_NC[999]
A_QC_9=meanA_QC[999]
s_QC_9=stdA_QC[999]
```
